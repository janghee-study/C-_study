```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen

url = 'https://sports.news.naver.com/kbaseball/record/index.nhn?category=kbo'

html = urlopen(url)
soup = BeautifulSoup(html, "html.parser")

soup
```

```python
soup.prettify
```

```python
soup.tbody.find_all('span')
```

```python
team = soup.tbody.find_all('span',id=True)
matchDetail = soup.body.tbody.find_all('span')
winloseRatio=soup.tbody.find_all('strong')
```

```python
#팀이름
teamname_list = [] #list()
for i in range(0,10):
    teamname_list.append(team[i].string)
    
teamname_list
```

```python
#경기수
matchCnt_list = list()

for i in range(1,100,10):
    matchCnt_list.append(matchDetail[i].string)
    
matchCnt_list
```

```python
#팀승리
win_list = list()

for i in range(2,100,10):
    win_list.append(matchDetail[i].string)
    
win_list
```

```python
#팀패배
lose_list = list()

for i in range(3,100,10):
    lose_list.append(matchDetail[i].string)
    
lose_list
```

```python
#팀승률
winRatio_list = list()

for i in range(1,20,2):
    winRatio_list.append(winloseRatio[i].string)
    
print(winRatio_list)
```

```python
html = open('KBO.html','w')

html.write('<body>')
html.write('''
    <table border=1>
        <tr>
            <th>순위</th>
            <th>팀명</th>
            <th>경기수</th>
            <th>승</th>
            <th>패</th>
            <th>승률</th>
        </tr>
''')
#teamname_list,matchCnt_list,win_list,lose_list,winRatio_list
for i in range(1,10):
    html.write('<tr>')
    html.write(f'<td>{i}</td>')
    html.write(f'<td>{teamname_list[i-1]}</td>')
    html.write(f'<td>{matchCnt_list[i-1]}</td>')
    html.write(f'<td>{win_list[i-1]}</td>')
    html.write(f'<td>{lose_list[i-1]}</td>')
    html.write(f'<td>{winRatio_list[i-1]}</td>')
    html.write('</tr>')

html.write('</body>')
html.write('</html>')
html.close()
```

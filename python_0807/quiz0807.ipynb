### 자원봉사자 연간 평균봉사시간

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
#csv파일을 긁어옴
#나중에 for문으로 변경
year2018 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2018.csv',encoding='utf-8')
year2018.drop([0,1],inplace=True)
year2017 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2017.csv',encoding='utf-8')
year2017.drop([0,1],inplace=True)
year2016 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2016.csv',encoding='utf-8')
year2016.drop([0,1],inplace=True)
year2015 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2015.csv',encoding='utf-8')
year2015.drop([0,1],inplace=True)
year2014 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2014.csv',encoding='utf-8')
year2014.drop([0,1],inplace=True)
year2013 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2013.csv',encoding='utf-8')
year2013.drop([0,1],inplace=True)
year2012 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2012.csv',encoding='utf-8')
year2012.drop([0,1],inplace=True)
year2011 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2011.csv',encoding='utf-8')
year2011.drop([0,1],inplace=True)
year2010 = pd.read_csv('../quiz0807/자원봉사자연간평균봉사시간2010.csv',encoding='utf-8')
year2010.drop([0,1],inplace=True)
```

```python
year2018.head()
```

```python
data_result = year2018
```

```python
#나중에 포문정렬
data_result = pd.merge(data_result,year2017,on='시도별')
data_result = pd.merge(data_result,year2016,on='시도별')
data_result = pd.merge(data_result,year2015,on='시도별')
data_result = pd.merge(data_result,year2014,on='시도별')
data_result = pd.merge(data_result,year2013,on='시도별')
data_result = pd.merge(data_result,year2012,on='시도별')
data_result = pd.merge(data_result,year2011,on='시도별')
data_result = pd.merge(data_result,year2010,on='시도별')
```

```python
data_result.head()
```

```python
data_result['2018']=data_result['2018'].astype('float64')
data_result['2017']=data_result['2017'].astype('float64')
data_result['2016']=data_result['2016'].astype('float64')
data_result['2015']=data_result['2015'].astype('float64')
data_result['2014']=data_result['2014'].astype('float64')
data_result['2013']=data_result['2013'].astype('float64')
data_result['2012']=data_result['2012'].astype('float64')
data_result['2011']=data_result['2011'].astype('float64')
data_result['2010']=data_result['2010'].astype('float64')
```

```python
# # data_result['8년 평균']=data_result['8년 평균'].astype('int64')
# data_result['8년 평균'] = (data_result['2018']+ data_result['2017'])
# #                         data_result['2017']+ data_result['2016']+ data_result['2015']+ data_result['2014']+ data_result['2013']+ data_result['2012']+ data_result['2011']+ data_result['2010']) / 8

# data_result
```

```python
data_result['AVG'] = (data_result['2017']+ data_result['2018']+ data_result['2016']+ data_result['2015']+ data_result['2014']+\
                    data_result['2013']+ data_result['2012']+ data_result['2011']+ data_result['2010']) /8
#소수점 보여주는 것
pd.options.display.float_format = '{:.1f}'.format
data_result1 = pd.DataFrame.copy(data_result) #pandas DF에서의 깊은복사방법
data_result
```

```python
#간격
t = np.arange(0,12,0.01)
y = np.sin(t)
#옵션
plt.figure(figsize=(7,2))
#피규어가 세팅이 되어야만 그리드를 사용할수있다.
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example')
#그리기
plt.plot(t,y)
plt.plot(t,np.cos(t))
plt.show()
```

```python
#간격
t = np.arange(0,12,0.01)
y = np.sin(t)
#옵션
plt.figure(figsize=(7,2))
#피규어가 세팅이 되어야만 그리드를 사용할수있다.
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example')
#그리기
plt.plot(t,y,label='sin')
plt.plot(t,np.cos(t),label='cos')
plt.legend() #타이틀보여줌
plt.show()
```

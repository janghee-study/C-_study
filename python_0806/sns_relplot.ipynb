### seaborn 연습

```python
#한글세팅
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':    #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
tips = sns.load_dataset("tips")
tips.head(7)
```

```python
#산정도 그리기
#g = sns.relplot(x="total_bill", y="tip", data=tips)
g = sns.relplot(x="total_bill", y="tip",hue='day',data=tips) 
#matplotlib와는 다르게 legend간편하다
```

```python
#columns을 통해서 컬럼구별가능!
g = sns.relplot(x="total_bill", y="tip", hue='day', col='time', data=tips)
```

```python
#row를 기준으로 성별로 나눴다!
g = sns.relplot(x="total_bill", y="tip", hue='day', col='time', row='sex', data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue='time', col='day',col_wrap=2, data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="time", size="size", 
                palette=["b", "r"], sizes=(10, 100), col="time", data=tips)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
sns.set(style="whitegrid")
```

```python
import matplotlib.font_manager as fm
import platform

if platform.system() =='Windows':
    font_path = 'C:/Windows/Fonts/gulim.ttc' #윈도우 굴림폰트
elif platform.system() == 'Linux':     #우분투 18.04 테스트됨, 나눔폰트 설치 후
    font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
        
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 11
```

```python
data1 = pd.read_excel('../data/연간고용률01.xlsx', encoding='utf-8')
data1.head()
```

```python
data1.연도.unique()
```

```python
#간격이 맞지않아서 수정이 불가피하다
plt.figure(figsize=(10,3))
sns.lineplot(x='연도',y='고용율',data=data1)
plt.show()
```

```python
import matplotlib.ticker as ticker
plt.figure(figsize=(10,4))
ax = sns.lineplot(x='연도',y='고용율',data=data1)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.show()
```

```python
x=data1.연도.tolist()
y=data1.고용율.tolist()
plt.figure(figsize=(15,4))
ax = plt.plot(x,y,'bo-',ms=8,mfc='r') # rgb로 색을 조절할수있다 bo- 조절할수있다.
plt.title('연도별 고용율 변화')
plt.xlabel('연도')
plt.ylabel('고용율')
plt.ylim(57.5,62.5)   # y축 보여지는 간격을 57.5~62.5로 하자!
plt.xticks(x)
plt.show()
#자세한것은 matplotlib 도큐먼트를 보고 확인하도록 하자.
#https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
```

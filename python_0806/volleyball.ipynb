```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
```

```python
driver.get("https://sports.news.naver.com/basketball/record/index.nhn?category=kbl")
```

```python
page = driver.page_source
```

```python
#table 정보추출모듈
from html_table_parser import parser_functions as parser
soup = BeautifulSoup(page,'html')
table = soup.find_all('tbody')
table = table[2]
```

```python
table2 = parser.make2d(table)
df = pd.DataFrame(table2)
```

```python
df.head()
```

```python
col_rename={0:'순위', 1:'선수명', 2:'경기수', 3:'AS', 4:'리바운드', 
            5:'스틸', 6:'블록', 7:'2점슛', 8:'3점슛', 
            9:'자유투', 10:'야투성공', 11:'3점슛성공', 12:'자유투성공', 
           }
df.rename(columns = col_rename, inplace=True)

df.set_index('순위',inplace=True)
df.head()
```

```python
import re
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen

url = 'https://finance.naver.com/sise/lastsearch2.nhn'

html = urlopen(url)
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
name = soup.find_all('a')
name
```

```python
day = soup.find_all(class_='number')
day
```

```python
name_list = list()

#if re.search('item',name[i].string) != none:
for i in range(60,84,1):
    name_list.append(name[i].string)
    
print(name_list)
```

```python
name[60].string
```

```python
stock_list = list()

for i in range(1,250,10):
    stock_list.append(day[i].string)
    
print(stock_list)
```

```python
html = open('stock.html','w')

html.write('<body>')
html.write('''
    <table border=1>
        <tr>
            <th>순위</th>
            <th>종목명</th>
            <th>현재가</th>
        </tr>
''')

for i in range(1,25):
    html.write('<tr>')
    html.write(f'<td>{i}</td>')
    html.write(f'<td>{name_list[i-1]}</td>')
    html.write(f'<td>{stock_list[i-1]}</td>')
    html.write('</tr>')

html.write('</body>')
html.write('</html>')
html.close()
```
